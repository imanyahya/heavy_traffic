# Finding Heavy Traffic Indicators on I-94
In this project, we are going to analyze a dataset about the westbound traffic on the [I-94 Interstate highway](https://en.wikipedia.org/wiki/Interstate_94). This dataset was made by John Hogue and can be downloaded from [UCI Machine Learning Repository](https://archive.ics.uci.edu/ml/datasets/Metro+Interstate+Traffic+Volume).

**Dataset Infomation:**  
Hourly Interstate 94 Westbound traffic volume for MN DoT ATR station 301, roughly midway between Minneapolis and St Pail, MN. Hourly weather features and holidays included for impacts on traffic volume.

**GOAL**: to determine a few indicators of heavy traffic on I-94. These indicators can be weather type, time of the day, time of the week, etc. For instance, we may find out that the traffic is usually heavier in the summer or when it snows.

*note: this exercise focused on data visualization, therefore we are using a clean dataset, and will go straight to the analysis process.*

Now, let's begin opening our dataset. We will store the dataset in `traffic` dataframe.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
traffic = pd.read_csv('Metro_Interstate_Traffic_Volume.csv')
traffic.head(3)
```

```python
traffic.tail(3)
```

```python
traffic.info()
```

As we can see, our `traffic` dataframe contains 48204 rows with 9 colums, and no null values. Based on the attribute information listed on the website, we have:
- `holiday` - Categorical US National holidays plus regional holiday, Minnesota State Fair
- `temp` - Numeric Average temp in kelvin
- `rain_1h` - Numeric Amount in mm of rain that occurred in the hour
- `snow_1h` - Numeric Amount in mm of snow that occurred in the hour
- `clouds_all` - Numeric Percentage of cloud cover
- `weather_main` - Categorical Short textual description of the current weather
- `weather_description` - Categorial Longer textual description of the current weather
- `date_time` - Date Time Hour of the data collected in local CST time
- `traffic_volume` - Numeric Hourly I-94 ATR 301 reported westbound traffic volume

## Analyzing Traffic Volume
First, we are going to plot a histogram to visualize the distribution of the `traffic_volume` column.

```python
traffic['traffic_volume'].plot.hist()
plt.xlabel('Traffic volume (cars)')
plt.ylabel('Frequency')
plt.title('Traffic Volume Histogram')
plt.show()
```

```python
traffic['traffic_volume'].describe()
```

Our data start from 2012-10-02 09:00:00 to 2018-09-30 23:00:00. During that duration, we have variation of traffic volumes between 0 to 7280 cars, with an average around 3260 cars.

Not much we can tell from the histogram other than there are a lot of occurrences where the traffic volume is around 0 ~ 500 cars and around 4500 ~ 5000 cars. The variations could be caused by several factors such as time of the day (low volumes most likely occur during the night), or weather condition.

In the next section, we will observe the relationship between traffic volume and time of the day.

## Traffic Volume: Day vs. Night

Previously, we analyzed the distribution of `traffic_volume`and find the following:
- About 25% of the time, there were 1,193 cars or fewer passing the station each hour - this probably occurs during the night, or when a road is under construction.
- About 25% of the time, the traffic volume was four times as much (4,933 cars or more).

This possibilty that nighttime and daytime might influence traffic volume gives our analysis an interesting direction: comparing daytime with nighttime data.

We will start by dividing the dataset into two parts:
- Daytime data: hours from 7 a.m. to 7 p.m.
- Nighttime data: hours from 7 p.m. to 7 a.m.

```python
traffic['date_time'] = pd.to_datetime(traffic['date_time'])
traffic['date_time'].dt.hour
```

```python
daytime = traffic.copy()[(traffic['date_time'].dt.hour >= 7) & 
                         (traffic['date_time'].dt.hour <= 19)]
nighttime = traffic.copy()[(traffic['date_time'].dt.hour < 7)|
                           (traffic['date_time'].dt.hour > 19)]
```

Now that we have separated datasets depending on the time of the day, we can compare them with the traffic volume.

```python
plt.figure(figsize=(15,6))

plt.subplot(1, 2, 1)
daytime['traffic_volume'].plot.hist()
plt.title('Daytime Traffic')
plt.xlabel('Traffic volume')

plt.subplot(1, 2, 2)
nighttime['traffic_volume'].plot.hist()
plt.title('Nighttime Traffic')
plt.xlabel('Traffic volume')

plt.xlim(0,7000)
plt.ylim(0,8000)
plt.show()
```

```python
daytime['traffic_volume'].describe()
```

Our histograms show two different characteristics, the daytime histogram is left skewed, meaning most of the data has high values for the traffic volume. On average, there are 4649 or more cars passing the station.

```python
nighttime['traffic_volume'].describe()
```

In contrast, the night time histogram is right skewed. Most of the traffic volumes recorded during the night has low values. While there is still some high values recorded (above 5000 cars), their occurrences are rare and generally the traffic during night time is low. Since our goal is to determine indicators on heavy traffic, we will disegard the night time data and will move forward only with the day time data.

## Time Indicators
One of the possible indicators of heavy traffic is time. There might be more people on the road in a certain month, on a certain day, or at a certain time of the day.

We're going to look at a few line plots showing how the traffic volume changed according to the following parameters:
- Month
- Day of the week
- Time of the day

### Traffic change each month

```python
daytime['month'] = daytime['date_time'].dt.month
by_month = daytime.groupby('month').mean()
by_month['traffic_volume']
```

```python
by_month['traffic_volume'].plot.line()
plt.title('Traffic Volumes by Month')
plt.xlabel('Month')
plt.ylabel('Traffic volume')
plt.show()
```

The traffic is generally high all year with exception between around November to February - possibly during winter less people use their personal vehicle, plus there are holidays. However we also observed low traffic volume around July, but we do not have any clear reason why it happened. So let's investigate further what happen during July.

```python
daytime['year'] = daytime['date_time'].dt.year
july = daytime[daytime['month'] == 7]
july_per_year = july.groupby('year').mean()

july_per_year['traffic_volume'].plot.line()
plt.title('Traffic Volume on July Each Year')
plt.xlabel('Year')
plt.ylabel('Traffic volume')
plt.show()
```

Actually, the traffic volume on july is generally as high as other months, but somehow the value was extremely low only in 2016. It is most likely an event occurred in that particular year, possibly a construction work.

After digging for a while, [this article](https://www.crainsdetroit.com/article/20160728/NEWS/160729841/weekend-construction-i-96-us-23-bridge-work-i-94-lane-closures-i-696) is most likely related to the low traffic volume on July 2016.

At this point, we can tetatively conclude that the traffic generally heavy during warm months (March - October).

### Traffic change each day of the week

```python
daytime['dayofweek'] = daytime['date_time'].dt.dayofweek
daily = daytime.groupby('dayofweek').mean()

daily['traffic_volume'].plot.line()
plt.title('Daily Traffic Volume')
plt.xlabel('Day')
plt.ylabel('Traffic volume')
plt.show()
```

The traffic is typically high during workday (Monday - Friday), then decreasing to under 4000 during weekends. This is reasonable since during weekends, people tends to stay at their homes.

The weekends, however, will drag down the average values, so we're going to look at the averages separately. To do that, we'll start by splitting the data based on the day type: business day or weekend.

```python
daytime['hour'] = daytime['date_time'].dt.hour

business_day = daytime.copy()[daytime['dayofweek'] <= 4]
weekend = daytime.copy()[daytime['dayofweek'] >= 5]

business_hour = business_day.groupby('hour').mean()
weekend_hour = weekend.groupby('hour').mean()

print(business_hour['traffic_volume'])
print(weekend_hour['traffic_volume'])
```

We have two series which inform us the traffic volume on each hour during business day and weekend separately. Let's plot each series so we can easily observe both series.

```python
plt.figure(figsize=(15,3))

plt.subplot(1, 3, 1)
business_hour['traffic_volume'].plot.line()
plt.title('Traffic during Business Days')
plt.ylabel('Traffic volume')
plt.xlabel('Hour')
plt.xlim(6,20)
plt.ylim(1500,7000)

plt.subplot(1, 3, 2)
weekend_hour['traffic_volume'].plot.line()
plt.title('Traffic during Weekend')
plt.ylabel('Traffic volume')
plt.xlabel('Hour')
plt.xlim(6,20)
plt.ylim(1500,7000)

plt.subplot(1, 3, 3)
business_hour['traffic_volume'].plot.line(label='Business days')
weekend_hour['traffic_volume'].plot.line(label='Weekend')
plt.title('Business Days vs. Weekend')
plt.ylabel('Traffic volume')
plt.xlabel('Hour')
plt.legend()
plt.xlim(6,20)
plt.ylim(1500,7000)

plt.show()
```

Average traffic volumes during business days is always higher than during weekend at every hour. The volumes are spiked at around 7.00 and 16.00 when people are go to work and back (rush hours). On weekend, traffic volumes increases at around 12 (possibly during lunch) but still lower than any time during business days.

So far we can conclude that:
- The traffic is generally heavier during warm months (March - October).
- The traffic is generally heavier during business days than weekend, especially during rush hours (at around 7.00 and 16.00).

## Weather Indicators
Another possible indicator of heavy traffic is weather. The dataset provides us with a few useful columns about weather: `temp`, `rain_1h`, `snow_1h`, `clouds_all`, `weather_main`, and `weather_description`.

A few of these columns are numerical so let's start by looking up their correlation values with `traffic_volume`.

```python
traffic.corr()['traffic_volume']
```

The highest correlation observed is on `temp` column with value around +0.13. The other relevant columns does not show any strong correlation with `traffic_volume`.

Let's try to visualize the correlation between `temp` and `traffic_volume` using scatter plot.

```python
plt.scatter(traffic['traffic_volume'], traffic['temp'])
plt.ylim(240, 320)
plt.show()
```

It seems there is no clear pattern displayed, so we can conclude that temperature is not a solid indicator of heavy traffic.

Now let's check at the non-numerical weather-related columns: `weather_main` and `weather_description`.

```python
weather_main = daytime.groupby('weather_main').mean()
weather_description = daytime.groupby('weather_description').mean()
```

```python
weather_main['traffic_volume'].sort_values().plot.barh()
plt.title('Traffic Volume vs. Weather')
plt.xlabel('Traffic volume')
plt.xlim(0,7000)
plt.show()
```

The traffic is relatively consistent across different types of weather. The traffic is heaviest during cloudy weather but not significant compared to other weather conditions, not to mention the traffic volume is still less than 5000 (around 4500). We can conclude that `weather_main` column is not strong indicator for traffic volume.

Next we will try to observe the more detailed weather type information from `weather_description` column.

```python
weather_description['traffic_volume'].sort_values().plot.barh(figsize=(10,20))
plt.title('Traffic Volume vs. Weather Desc')
plt.xlabel('Traffic volume')
plt.xlim(0,7000)
plt.show()
```

We have three weather condition where the traffic volumes exceed 5000 cars:
- Shower snow
- Light rain and snow
- Proximity thunderstorm with drizzle

We are not sure why there is high traffic volume during those three weather conditions. Those conditions are not good but not that bad either, just in the middle (most likely more people use their personal vehicle). Probably during clear or slightly bad weather, people still prefer to walk, bike, or use public transportation, while during very bad weather people prefer to take a leave and stay at home.
